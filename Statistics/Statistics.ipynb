**Descriptive Statitics**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Simulated house prices dataset (in thousands of dollars)
house_prices = [150, 200, 250, 300, 350, 400, 200, 250, 275, 325, 300, 400, 500, 150, 180]

# Descriptive statistics
mean_price = np.mean(house_prices)
median_price = np.median(house_prices)
mode_price = pd.Series(house_prices).mode()[0]
std_dev_price = np.std(house_prices)
variance_price = np.var(house_prices)
range_price = max(house_prices) - min(house_prices)

print("House Prices Descriptive Statistics:")
print(f"Mean Price: {mean_price:.2f}")
print(f"Median Price: {median_price}")
print(f"Mode Price: {mode_price}")
print(f"Standard Deviation: {std_dev_price:.2f}")
print(f"Variance: {variance_price:.2f}")
print(f"Range: {range_price}")

# Visualizing the house price distribution
sns.histplot(house_prices, bins=8, kde=True, color="green")
plt.title("House Price Distribution")
plt.xlabel("Price (in $1000s)")
plt.ylabel("Frequency")
plt.show()
```

**Inferential Statistics**

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Sample data: exam scores of 30 students
exam_scores = [65, 72, 80, 68, 75, 77, 82, 66, 70, 71,
               79, 85, 71, 73, 76, 70, 78, 74, 69, 68,
               80, 83, 67, 72, 74, 75, 76, 78, 72, 70]

# Step 1: Hypothesis Testing (One-sample t-test)
# Null Hypothesis (H0): The average score is 70
# Alternative Hypothesis (H1): The average score is not 70

population_mean = 70
sample_mean = np.mean(exam_scores)
sample_size = len(exam_scores)
sample_std = np.std(exam_scores, ddof=1)  # Sample standard deviation

# T-statistic for one-sample t-test
t_stat, p_value = stats.ttest_1samp(exam_scores, population_mean)

print(f"Sample Mean: {sample_mean:.2f}")
print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.4f}")

# Decision rule: If p-value < 0.05, reject H0.
if p_value < 0.05:
    print("We reject the null hypothesis: The average score is significantly different from 70.")
else:
    print("We fail to reject the null hypothesis: There is no significant difference from 70.")

# Step 2: Confidence Interval (95% Confidence Interval for the Mean)
confidence_level = 0.95
standard_error = sample_std / np.sqrt(sample_size)

# t-critical value for 95% confidence and degrees of freedom (sample_size - 1)
t_critical = stats.t.ppf((1 + confidence_level) / 2, df=sample_size - 1)

# Confidence interval calculation
margin_of_error = t_critical * standard_error
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print(f"95% Confidence Interval for the Mean: ({confidence_interval[0]:.2f}, {confidence_interval[1]:.2f})")

# Step 3: Visualization
plt.figure(figsize=(10, 6))
sns.histplot(exam_scores, bins=8, kde=True, color='skyblue', edgecolor='black')
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=1.5, label=f"Sample Mean: {sample_mean:.2f}")
plt.axvline(confidence_interval[0], color='green', linestyle='dashed', linewidth=1.5, label=f"Lower CI: {confidence_interval[0]:.2f}")
plt.axvline(confidence_interval[1], color='orange', linestyle='dashed', linewidth=1.5, label=f"Upper CI: {confidence_interval[1]:.2f}")
plt.title("Distribution of Exam Scores with Confidence Interval")
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

**Probability**

```python
# Probability Distributions
from scipy.stats import binom, norm

# Binomial Distribution
n, p = 10, 0.5
x = range(0, n + 1)
binom_probs = binom.pmf(x, n, p)

plt.bar(x, binom_probs, color='purple')
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.show()

print()
# Normal Distribution
x = np.linspace(-4, 4, 100)
y = norm.pdf(x, 0, 1)

plt.plot(x, y, color='green')
plt.title("Standard Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.show()
```

**Frequency Distribution**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset: Exam scores of 30 students
exam_scores = [65, 72, 80, 68, 75, 77, 82, 66, 70, 71,
               79, 85, 71, 73, 76, 70, 78, 74, 69, 68,
               80, 83, 67, 72, 74, 75, 76, 78, 72, 70]

# Create a pandas Series from the dataset
data = pd.Series(exam_scores)

# Absolute Frequency
absolute_freq = data.value_counts().sort_index()
print("Absolute Frequency:\n", absolute_freq)

# Relative Frequency
relative_freq = data.value_counts(normalize=True).sort_index()
print("\nRelative Frequency:\n", relative_freq)

# Visualization

# Bar Chart for Absolute Frequency
plt.figure(figsize=(10, 6))
absolute_freq.plot(kind='bar', color='skyblue')
plt.title('Bar Chart: Absolute Frequency Distribution')
plt.xlabel('Exam Score')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()

print()
# Histogram for the Frequency Distribution (for continuous data)
plt.figure(figsize=(10, 6))
plt.hist(exam_scores, bins=10, color='lightgreen', edgecolor='black')
plt.title('Histogram: Frequency Distribution of Exam Scores')
plt.xlabel('Exam Score')
plt.ylabel('Frequency')
plt.show()
```
